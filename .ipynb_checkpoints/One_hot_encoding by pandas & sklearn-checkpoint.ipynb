```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
data=sns.load_dataset('tips')
data.head()
```

## get_dummies() function is belongs to pandas library so it is work only on pd.get_dummies() not dataframe like df.get_dummies()

```python
# all categorical columns divided
# and create a numerical columns for each class 
df2=pd.get_dummies(data)
df2.head()
```

```python
data['day'].value_counts()
```

# prefix parameter

```python
pd.get_dummies(data,prefix='col').head(1)              #using prefix parameter
# prefix_sep="-" by default underscor you can change it
```

# prefix with prefix_sep parameter

```python
pd.get_dummies(data,prefix='columns',prefix_sep='-->')[['columns-->Thur']].head(2)   # you can see now 
```

# columns parameter

```python
pd.get_dummies(data,columns=['sex','time']).head(2)
```

## dummy_na =True is used for missing value if value is present than it will put 1 otherwise 0 , when will you use dummy_na parameter than it will create two columns

```python
pd.get_dummies(data,prefix='column',prefix_sep='--',columns=['day','time'],dummy_na=True)['column--nan'].head()   #by defaul dummy_na is False 
# but you can use 
```

# By usign sklearn library

```python
from sklearn.preprocessing import OneHotEncoder
```

# sparse =True parmeter

```python
#sparse=True    ----->   it will return metrics
#sparse=False   ----->   it will return ndarray
on_endr=OneHotEncoder(sparse=True)
```

```python
b=on_endr.fit_transform(data[['day']])
```

```python
print(b)
```

# sparse =False parmeter

```python
obj2=OneHotEncoder(sparse=False)
```

```python
obj2.fit_transform(data[['day']])[0:5,0:]
```

```python
# from sklearn.preprocessing import OneHotEncoder
obj3=OneHotEncoder(sparse=False,dtype='int64')
```

```python
obj3.fit_transform(data[['day','time','sex']]).shape
```

```python
obj3.fit_transform(data[['day','time','sex']])
```

```python
arrayy=obj3.fit_transform(data[['day','time','sex']])
```

# update in my dataframe

```python
pd.get_dummies(data,columns=['day','time','sex']).keys()
```

```python
df=pd.DataFrame(arrayy,columns=['day_Thur', 'day_Fri', 'day_Sat',
       'day_Sun', 'time_Lunch', 'time_Dinner', 'sex_Male', 'sex_Female'])
```

```python
df.head()
```
